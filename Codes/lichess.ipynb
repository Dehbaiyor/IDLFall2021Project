```python
import pandas as pd
import re
```

```python
chess_data = pd.read_csv('../Data/lichess-08-2014.csv', delimiter = ',')
chess_data = chess_data[chess_data['Mode'] != 'Bullet']
```

```python
chess_data.head()
```

```python
chess_data['result'] = chess_data.Result.map({'Black Wins': '0-1', 'White Wins': '1-0', 'Draw':'1/2-1/2'})
chess_data['game'] = chess_data['PGN'] + ' ' + chess_data['result']
chess_data
```

```python
chess_data['game'][0]
```

```python
def clean(strng):
    return re.sub(r'[0-9]+\.', '', strng).lstrip().replace('  ',' ')
```

```python
clean(chess_data['game'][0])
```

```python
chess_data['PGN'] = chess_data['game'].apply(clean)
```

```python
chess_data['PGN'].to_csv('../Data/cleaned_lichess08.csv', index = False)
```
