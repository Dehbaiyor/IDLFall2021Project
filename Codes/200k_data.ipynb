```python
import pandas as pd
import re
```

```python
chess_data = pd.read_csv('../Data/200k_blitz_rapid_classical_bullet.csv', delimiter = ',').fillna('')
#chess_data = chess_data[chess_data['Mode'] != 'Bullet']
```

```python
moves_cols = []
eval_cols = []

for i in range(1,201):
    moves_cols.append('Move_ply_' + str(i))
    eval_cols.append('Eval_ply_' + str(i))
    
moves_cols.append('Result')
```

```python
chess_data['Moves'] = chess_data[moves_cols].astype(str).apply(' '.join, axis=1)
```

```python
chess_data['Evals'] = chess_data[eval_cols].astype(str).apply(' '.join, axis=1)
```

```python
chess_data[['Moves', 'Evals']]
```

```python
clean(chess_data['Evals'][0])
```

```python
def clean(strng):
    return re.sub(' +', ' ', strng)
```

```python
chess_data['PGN'] = chess_data['Moves'].apply(clean)
```

```python
chess_data['Evals'] = chess_data['Evals'].apply(clean)
```

```python
chess_data['game_length'] = chess_data['PGN'].str.split(' ').apply(len)
chess_data['eval_length'] = chess_data['Evals'].str.split(' ').apply(len)
```

```python
chess_data[['game_length','eval_length']]
```

```python
chess_data[['PGN', 'Evals']].to_csv('../Data/cleaned_200k.csv', index = False)
```
