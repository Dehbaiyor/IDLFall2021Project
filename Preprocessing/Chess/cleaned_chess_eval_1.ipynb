### Data processing

```python
eval_df = pd.DataFrame(columns=["PGN", "Evals"])
chunksize = 10 ** 6

i = 0
with pd.read_csv("../data/cleaned_merged_chess_data.csv", chunksize=chunksize) as reader:
    for chunk in reader:
        if not chunk.dropna(subset=['Evals']).empty:
            eval_df = eval_df.append(chunk.dropna(subset=['Evals']), ignore_index=True)
        i += 1
        print(i)
```

```python
def t5_format(df, columns=["Move", "Eval"]):
    new_df = pd.DataFrame(columns=columns)
    count = 0
    for index in range(len(df)):
        moves, evals = df.iloc[index][0].split(), df.iloc[index][1].split()
        for i in range(len(evals)):
            new_row = {columns[0]:" ".join(moves[:i+1]), columns[1]:evals[i]}
            new_df = new_df.append(new_row, ignore_index=True)
        count += 1
        print("-"*30)
        print("#"*5, count, len(new_df))
        print(new_df)
    return new_df

def t5_format2(array):
    out = []
    for arr in array:
        moves, evals = arr[0].split(), arr[1].split()
        new_arr = np.zeros(shape=(len(evals), 2), dtype=object)
        for i in range(new_arr.shape[0]):
            new_arr[i][0] = " ".join(moves[:i+1])
            new_arr[i][1] = evals[i]
        out.extend(new_arr)
    return out

def my_func(arr):
    arr = arr[0]
    print("-"*50)
    moves, evals = arr[0].split(), arr[1].split()
    new_arr = np.zeros(shape=(len(evals), 2), dtype=object)
    for i in range(new_arr.shape[0]):
        new_arr[i][0] = " ".join(moves[:i+1])
        new_arr[i][1] = evals[i]
    print(arr.shape)
    return new_arr
```
