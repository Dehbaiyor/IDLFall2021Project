```python
def create_sequences(game):
    seq = []
    game = game.split(' ')
    if len(game) <= 1:
        return [['e4 : ', 'e4']]
    choices = np.random.randint(0, len(game) - 1, 10)
    
    for i in choices: #range(1, len(game) - 1):
        inp = 'e4' + ' : ' + ' '.join(game[:i])
        outp = ' '.join(game[i:])        
        seq += [[inp, outp]]
    return seq
```

```python
from random import shuffle

with pd.read_csv('lichess_elite.csv', 
                 chunksize=100_000) as reader:
    for num, chunk in enumerate(reader):
        print(f'Chunk {num + 1} of 40')
        finetuned_data = []
        for game in tqdm(chunk['PGN']):
            finetuned_data += create_sequences(game)
        shuffle(finetuned_data)
        pd.DataFrame(finetuned_data, 
                    columns = ['input', 'target']).to_csv(f'chess_data/{num + 1}.csv', index = False)
```
