```python
import pandas as pd
df = pd.read_csv('../input/idl-project-code-encoder-set/encoder_code_to_docstring_test.csv')
```

```python
ai3 = aitextgen(model_folder="../input/gpt-finetuned-model/gpt_code_task2_modelv1/gpt_code_task2_modelv1",
               tokenizer_file="../input/gpt-finetuned-model/gpt_code_task2_modelv1/gpt_code_task2_modelv1/aitextgen.tokenizer.json",
               to_gpu=True)
```

```python
i = 200
string = df.iloc[i]['input']
string[:-28], string[-28:]
```

```python
p = 'summarize: def sharpe_ratio(return_series, N = 255, rf = 0.01, annualized = True):\
    mean = return_series.mean() * N -rf\
    sigma = return_series.std()\
    if annualized:\
        sigma *= np.sqrt(N)\
    return mean / sigma'
```

```python
%%time
a  = ai3.generate(max_length = 512, 
            prompt = p, 
            top_k = 100, 
            top_p  = 0.9, 
            temperature = 0.7,
            return_as_list = True,
            n = 10)
```

```python
for a_ in a:
    print(a_.replace(p,""))
```
